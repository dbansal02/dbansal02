```python
import numpy as np
import matplotlib.pyplot as plt

def diffusion_equation(x, t, D):
    norms = np.linalg.norm(x, axis=0)  # Calculate norm along columns (axis=0)
    return 1 / (4 * np.pi * D * t) * np.exp(-norms**2 / (4 * D * t))

# Assuming X_flat and T_flat are defined somewhere
N = 100
X_train = np.zeros((N, 2))  # Assuming 2D space
X_train[0:N, :] = np.vstack((X1_flat, X2_flat)).T
print("X_train:")
print(X_train)
D = 0.8
y = diffusion_equation(X_train.T, 1, D)

# Assuming y_train is a 1D array
y_train = y.reshape(-1, 1)

print("\ny_train:")
print(y_train)

fig, ax = plt.subplots(figsize=(8, 8), dpi=600)
print(y_train.shape)
y_train = y.reshape(N, 1)
print(y_train.shape)
contour = ax.contourf(X1_flat, X2_flat, y_train.reshape(X1_flat.shape), cmap='coolwarm')
ax.set_xlabel('x1')
ax.set_ylabel('x2')
ax.set_title('Contour for Predicted Y for t=1')
plt.colorbar(contour, label='Predicted Y')

plt.savefig('predicted_Y_contour.png', bbox_inches='tight')
plt.show()
```
